Create a dataframe from a dictionary and list.

```python
data1 = {'col_1': [3, 2, 1, 0], 'col_2': ['a', 'b', 'c', 'd'], 'col_3': [19, 4, 7, 10],'col_4': ['aa', 'bb', 'cc', 'dd']}
data2 = [[3, 'a', 19, 'yes'], [2, 'b', 4, 'no'], [1, 'c', 7, 'yes'], [0, 'd', 10, 'no']]
```

```python
a1 = pd.DataFrame.from_dict(data1)
a2 = pd.DataFrame.from_dict(data1, orient='index')
a3 = pd.DataFrame.from_dict(data1, orient='index', columns=['A', 'B', 'C', 'D'])
a4 = pd.DataFrame(data2, columns=['A', 'B', 'C', 'D']) 
```

```python
df = a4
print('First row : ', list(df.loc[0]))
print('\nFirst and third row : \n', df.loc[[0, 2]])
print('\nRow with index 2 as a frame : \n', df.loc[2])
print('\nRow with index 2 as a list : \n', df.loc[[2]])
print('\nRow with index 2 and column D : \n', df.loc[2, 'D'])
print('\nRows with A column > 1 : \n', df.loc[df['A']>1])
print('\nRows 0 to 1 and columns A and B :\n',df.loc[df.index[0:2],["A","B"]])
```

Null Values, dropna, fillna, duplicates
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html

```python
df = a4.copy(deep = True)
df.iloc[1, 2] = None
df
```

```python
b = df.dropna()
c = df.dropna(axis='columns')
d = df.dropna(how='all')
e = df.dropna(subset=['A', 'B'])
f = df.fillna('New Value')
g = df.isna()
h = df.isnull()
i = df['C'].fillna(df["C"].mean())
print('Row with null value removed : \n', b)
print('\nColumn with null value removed : \n', c)
print('\nRow having all values are null removed : \n', d)
print('\nCheck only the columns A and B for null values :\n', e)
print('\nChange the null value to new value : \n', list(f.iloc[1]))
print('\nMissing values are detected : \n', g)  # Reverse is df.notnull() : detects not null values.
print('\nThat''s the same with isna : \n', h)
print('\nChange the null value to average of the column C :\n',i)
print('\nNumber of duplicated rows :\n',df.duplicated().sum())
print('\nRows with not null values in column C :\n',df[df.C.notnull()])
```

### Aggregation

Count, Mean, Median, Mode, Sum, Count, Standard Deviation, Variation, Minimum, Maximum, GroupBy

```python
df = a4.copy(deep = True)
df
```

```python
df.describe()
```

```python
a = df['A'].count()
b = df['A'].mean()
c = df['A'].median()
d = df['A'].std()
e = df['A'].var()
f = df['A'].sum()
g = df['A'].prod()
h = df['A'].min()
i = df['A'].max()
j = df['A'].head(1)
k = df['A'].tail(1)
l = df['A'].mode()
m = df.groupby('D').sum()
n = df.groupby('D')['A'].sum()
o = df.agg(['sum', 'mean', 'max'])
p = df.agg({'A' : ['sum', 'min'], 'C' : ['min', 'max']})
print('Number of rows in column A :\n', a)
print('\nAverage of values in column A : \n',b)
print('\nMedian in column A : \n',c)
print('\nStandard deviation in column A : \n',d)
print('\nVariance in column A : \n',e)
print('\nSum of values in column A : \n',f)
print('\nProduct of all items  in column A : \n',g)
print('\nMinimum value in column A : \n',h)
print('\nMaximum value in column A : \n',i)
print('\nFirst value in column A : \n',j)
print('\nLast value in column A : \n',k)
print('\nSum of columns after grouped by yes or no : \n',m)
print('\nSum of column A by yes or no : \n',n)
print('\nUsing agg function to find sum, mean and max : \n',o)
print('\nUsing agg function to find sum, min of A and min, max of C : \n',p)
```

### Boolean Selection

```python
df = a4.copy(deep = True)
df
```

```python
a = df['C'] > 5
b = df[df['C'] > 5]
c = df[(df['C'] == 4) | (df['C'] == 7)]
d = df[df['C'].isin([4, 7])]
e = df[df['C'].between(1, 10)]    #If you want to exclude boundries, use inclusive = False
f = df[df['D'].str.contains('ye')]
g = df[df['D'].str.startswith('n')]    #There is also endswith
h = df.query('C == 7 & D == "yes"')
print('Column C if greater than 5 or not :\n',list(a))
print('\nRows with Column C greater than 5 :\n',b)
print('\nRows with Column C equal to 4 or 7 :\n',c)
print('\nRows with Column C equal to 4 or 7 (using isin) :\n',d)
print('\nRows with Column C between 1 and 10 (using between) :\n',e)
print('\nRows with Column D containig substring ye :\n',f)
print('\nRows with Column D containig starting with n :\n',g)  
print('\nRows with Column C is 7 and column D is yes with query function :\n',h)
```

### Operations on dataframes

```python
df = a4.copy(deep = True)
df
```

```python
a = df.C + 5     #or df['C'].add(5)
b = df.C - 5     #or df['C'].sub(5)
c = df.C * 5     #or df['C'].mul(5)
d = df.C / 5     #or df['C'].div(5)
e = df['C'] // df['A']    #or df['C'].floordiv(df['A'])
f = df['C'] % df['A']     # or df['C'].mod(df['A'])
g = df['C'] ** df['A']    # or df['C'].pow(df['A'])
print('Add 5 to the values of column C :\n', list(a))
print('\nSubtract 5 from the values of column C :\n', list(b))
print('\nMultiply the values of column C by 5 :\n', list(c))
print('\nDivide the values of column C by 5 :\n', list(d))
print('\nDivide the column C by A, then round down :\n', list(e))
print('\nColumn C mode column A :\n', list(f))
print('\nColumn C to the power of column A :\n', list(g))
```
